```python
data = df.groupby(['group','age']).count()[['movie']].reset_index()
# for group
for group in data.group.unique():
    plt.plot(data[data.group == group].age, data[data.group == group].movie, label=group)
plt.legend()
```

```python
movies = [movie for movie in df.groupby('movie').count()['age'].sort_values()[::-1].index]
plt.figure(figsize=(25,15))
plt.rcParams.update({'font.size': 22})

plt.xticks(rotation=90,)
for group in data.group.unique():
    l = [
        data[(data.movie==movie)&(data.group==group) ]['age'] 
        if len(data[(data.movie==movie)&(data.group==group) ]['age']) > 0
        else 0
        for movie in movies
    ]
    plt.plot(movies, l,label=group)
plt.legend()
```
